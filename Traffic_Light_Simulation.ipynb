```python
import time

RED = "\033[91m‚óè\033[0m"
GREEN = "\033[92m‚óè\033[0m"
YELLOW = "\033[93m‚óè\033[0m"

def show(red=False, yellow=False, green=False):
    print(f"RED:    {RED if red else '‚óã'}")
    print(f"YELLOW: {YELLOW if yellow else '‚óã'}")
    print(f"GREEN:  {GREEN if green else '‚óã'}")
    print("-" * 20)

def traffic_light():
    while True:
          show(red=True)
          time.sleep(3)

          show(green=True)
          time.sleep(3)

          show(yellow=True)
          time.sleep(1)

traffic_light()
```

```python
import time
from IPython.display import clear_output

def draw(red=False, yellow=False, green=False):
    clear_output(wait=True)

    r = "üî¥" if red else "‚ö´"
    y = "üü°" if yellow else "‚ö´"
    g = "üü¢" if green else "‚ö´"

    print("Traffic Light Simulation/n")
    print(r)
    print(y)
    print(g)

def traffic_light():
    while True:
         draw(red=True)
         time.sleep(2)

         draw(green=True)
         time.sleep(2)

         draw(yellow=True)
         time.sleep(1)

traffic_light()
```
